# 모델링 및 모델 해석
문제 유형에 맞는 후보 모델군 선정, 모델군에서 최적 모델(다양한 모델 사용 후 성능비교 후 선정)을 학습 후 성능 확인, 모델 해석을 위한 Feature importances, ice, pdp 및 시각화 기법으로 모델의 작동을 해석(해석 유의사항 준수)

- baseline 모델을 선정하고, 모델링 성능 평가(모델 fit > baseline)
- feature engineering을 반복하여 최적의 모델 성능 찾기(모델 성능 개선 적용 방법)
- 문제정의에 맞는 (여러) 평가지표로 모델 성능 설명
- cv/hold-out으로 학습과 일반화(test-set 사용 권장) 성능 구분하여 제시(최종 모델과 일반화 성능을 설명)
- permutation importance, pdp 등으로 최종모델 설명(시각화는 설명이 가장 중요함)
- 제시한 가설 해결
